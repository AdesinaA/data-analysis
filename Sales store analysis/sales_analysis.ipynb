```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
sales = pd.read_csv("Downloads/SampleSuperstore.csv")
```

```python
sales.head()
```

```python
sales.shape
```

```python
sales.info()
```

```python
#find the number of missing values for all columns
sales.isnull().sum()
```

```python
#find the number of duplicate data
sales.duplicated().sum()
```

```python
#show the duplicate rows
sales[sales.duplicated(keep = 'last')]
```

```python
#drop the duplicated rows
sales.drop_duplicates(inplace = True)
```

```python
sales.shape
```

```python
#create a calculated field for profit margin
sales['profit Margin %'] = (sales.Profit / sales. Sales) * 100
sales.head()
```

```python
#get descriptive stats summary
sales.describe(include = "all")
```

```python
#group sales, profit and quantity by category [which category is best selliing & most profitable]
category_analysis = pd.DataFrame(sales.groupby(['Category'])[['Sales', 'Profit', 'Quantity']].sum())
category_analysis
```

```python
#set for grouped data
sns.set_theme(style="whitegrid")
figure, axis =  plt.subplots(1,3, figsize= (8,5))

#plot barplots
cat1 = sns.barplot(x = category_analysis.index, y =  category_analysis.Sales, ax = axis[0])
cat2 = sns.barplot(x = category_analysis.index,y = category_analysis.Profit, ax =  axis[1])
cat3 = sns.barplot(x = category_analysis.index, y =  category_analysis.Quantity, ax = axis[2])

#set titles
cat1.set(title = 'Sales')
cat2.set(title = 'Profit')
cat3.set(title = 'Quantity')

#rotate axis for x-axis
plt.setp(cat1.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat2.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat3.get_xticklabels(), rotation = 'vertical', size = 9)

#set spacing between subplots
figure.tight_layout()
```

```python
# [ what are the best selling & most profitable sub category ]
subcat_analysis =  pd.DataFrame(sales.groupby(['Sub-Category'])[['Sales', 'Profit']].sum())
```

```python
#sort by descending order according to sales
subcat_sales = pd.DataFrame(subcat_analysis.sort_values('Sales', ascending =  False))
subcat_sales
```

```python
# sort by descending order according to profit
subcat_profit = pd.DataFrame(subcat_analysis.sort_values('Profit', ascending = False))
subcat_profit
```

```python
#plot bar plots
sns.set_theme(style='whitegrid')

#set for grouped plots
figure, axis = plt.subplots(1,2, figsize=(12,6))

#  plot  bar plot for best selling sub category
subcat1 = sns.barplot(data = subcat_sales, x = subcat_sales.index, y = subcat_sales.Sales, ax = axis[0])
subcat1.set(title = 'Best Selling Sub-Category')
subcat1.set_xticklabels(subcat1.get_xticklabels(), rotation = 'vertical', size = 10)

# plot bar plot for most profitable sub category
subcat2 = sns.barplot(data = subcat_profit, x = subcat_profit.index, y = subcat_profit.Profit, ax = axis[1])
subcat2.set(title = 'Most Profitable Sub-Category')
subcat2.set_xticklabels(subcat1.get_xticklabels(), rotation = 'vertical', size = 10)


# set spacing between subplots
figure.tight_layout()

plt.show()
```

```python
# which is the best selling sub-category?
subcat_quantity =  pd.DataFrame(sales.groupby(['Sub-Category'])[['Quantity']].sum().sort_values('Quantity', ascending=False))
subcat_quantity
```

```python
# plot bar plot for top selling sub-category
sns.set_theme(style='whitegrid')
sns.barplot(data = subcat_quantity, y = subcat_quantity.index, x = subcat_quantity.Quantity, palette = 'muted')
plt.title('Top Selling Sub-Category')
plt.show()
```

```python
# which customer segment is most profitable
segment_analysis = pd.DataFrame(sales.groupby(['Segment'])[['Profit']].sum())
segment_analysis
```

```python
# plot bar plot
sns.set_theme(style='whitegrid')
sns.barplot(data = segment_analysis, x = segment_analysis.index, y = segment_analysis.Profit, palette = 'rocket')
plt.title('Customer segment profitability')
plt.show()
```

```python
# which is the preferred ship mode
sns.set_theme(style='whitegrid')
sns.countplot(sales['Ship Mode'])
plt.title('Ship Mode')
plt.show()
```

```python
region_analysis = pd.DataFrame(sales.groupby(['Region'])['Profit'].sum().reset_index())
region_analysis
```

```python
# plot piechart
explode = [0,0,0,0.1]

plt.pie(region_analysis.Profit,labels = region_analysis.Region, startangle = 90, autopct = '%1.0f', explode = explode, shadow = True)
plt.title('Most Profitable by Region')

plt.show()
```

```python
# which city has the highest number of sales
city_sales = pd.DataFrame(sales.groupby(['City'])[['Sales', 'Quantity']].sum().sort_values('Sales', ascending=False))
top15 = city_sales[:15]
top15
```

```python
bottom15 = city_sales[-15:]
bottom15
```

```python
# set for grouped  plots
figure, axis = plt.subplots(1,2,figsize=(12,5))

sns.set_theme(style='whitegrid')

top15c = sns.barplot(data = top15, y= top15.index, x = top15.Sales, palette = 'coolwarm', ax = axis[0])
top15c.set(title = 'Top 15 cities with highest sales')
top15c.set_yticklabels(top15c.get_yticklabels(), size =  10)

# plot bar plot for best selling sub category
bottom15c = sns.barplot(data = bottom15, y = bottom15.index, x =  bottom15.Sales, palette = 'coolwarm', ax= axis[1])
bottom15c.set(title= 'Bottom 15 cities with lowest sales')
bottom15c.set_yticklabels(bottom15c.get_yticklabels(), size = 10)

# set spacing between subplots
figure.tight_layout()

plt.show()
```
