```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# create color array for red dataframe
colour_red = np.repeat('red',red.shape[0])

# create color array for white dataframe
colour_white  = np.repeat('white', white.shape[0])
```

```python
# add red array to red
red['color'] = colour_red
red.head()
```

```python
# add white array to white
white['white'] = colour_white
white.head()
```

```python
# rename columns with space with _in red
red.rename(columns=lambda x: x.strip().lower().replace(' ','_'), inplace = True)
```

```python
red.head()
```

```python
# # rename columns with space with _in white
white.rename(columns=lambda x: x.strip().lower().replace(' ', '_'), inplace = True)
```

```python
white.head()
```

```python
# rename column named white to color
white.rename(columns={'white':'color'}, inplace = True)
```

```python
white.head()
```

```python
# rename all the columns in red to _ from __ to match with the white dataset
red.rename(columns={'fixed__acidity':'fixed_acidity'},inplace = True)
red.rename(columns={'volatile__acidity':'volatile_acidity'}, inplace = True)
red.rename(columns={'citric__acid':'citric_acid'}, inplace = True)
red.rename(columns={'residual__sugar':'residual_sugar'},inplace = True)
red.rename(columns={'free__sulfur__dioxide':'free_sulfur_dioxide'}, inplace = True)
red.rename(columns={'total__sulfur__dioxide':'total_sulfur_dioxide'}, inplace =  True)
```

```python
# concat the dataframes
liquor_df = pd.concat([red, white], ignore_index=True)
```

```python
# plot heatmap for feature variables
numeric_df = liquor_df.select_dtypes(include='number')
corr = numeric_df.corr()
plt.figure(figsize=[10,8])
sns.heatmap(corr, annot=True)
plt.xticks(rotation=45)
plt.title('heatmap of correlation coefficient for wine feature variables', size = 12);
```

```python
# view the min, 25%, 50%, 75% & max values with pandas describe
liquor_df['ph'].describe()
```

```python
# find the mean quality of each liquor type (red & white) with groupby
liquor_df.groupby('color')['quality'].mean()
```

```python
# bin edges that will be used to cut the data into groups
bin_edges = [2.72, 3.11, 3.21, 3.32, 4.01]
```

```python
# labels for the four acidity level groups
bin_names = ['high', 'moderately_high', 'medium', 'low']
```

```python
# creates acidity levels column
liquor_df['acidity_levels'] = pd.cut(liquor_df['ph'],bins=bin_edges,labels=bin_names)

liquor_df.head()
```

```python
# find the mean quality of each activity level with groupby
liquor_df.groupby('acidity_levels')['quality'].mean()
```

```python
# get the median amount of alcohol content
liquor_df['alcohol'].median()
```

```python
# select samples with alcohol content less than the median
low_alcohol = liquor_df.query('alcohol < 10.3')

# select samples with alcohol content greater than or equal to the median
high_alcohol = liquor_df.query('alcohol >= 10.3')

#ensure these queries include each sample exactly once
num_samples = liquor_df.shape[0]
num_samples == liquor_df['quality'].count() + high_alcohol['quality'].count()
```

```python
# get mean quality rating for the low alcohol group
high_alcohol['quality'].mean()
```

```python
# get mean quality rating for the high alcohol group
liquor_df['residual_sugar'].median()
```

```python
# select samples with residual sugar less than the median
low_sugar = liquor_df.query('residual_sugar < 3.0')

# select samples with residual sugar greater than or equal to the median
high_sugar  = liquor_df.query('residual_sugar >= 3.0')

#ensure these queries include each sample exactly once
num_samples == low_sugar['quality'].count() + high_sugar['quality'].count()
```

```python
# get mean quality rating for the low sugar group
low_sugar['quality'].mean()
```

```python
# get mean quality rating for the high sugar group
high_sugar['quality'].mean()
```

```python
# use query to select each group and get it's own quality
median = liquor_df['alcohol'].median()

# get subsets for data with < and >= alcohol median
low = liquor_df.query('alcohol < {}'.format(median))
high = liquor_df.query('alcohol >=  {}'.format(median))

# get mean for quality with low & high alcohol
mean_quality_low_alcohol = low['quality'].mean()
mean_quality_high_alcohol = high['quality'].mean()
```

```python
# create a bar chart with proper labels
locations = [1,2]
heights = [mean_quality_low_alcohol, mean_quality_high_alcohol]
labels = ['Low', 'High']

plt.bar(locations, heights, tick_label=labels)
plt.title('Average quality rating by alcohol content')
plt.xlabel('Alcohol content')
plt.ylabel('Average quality rating');
```

```python
# use query to select each group and get it's mean quality
median =  liquor_df['residual_sugar'].median()

# get the subsets for < and >= median
low_sugar = liquor_df.query('residual_sugar < {}'.format(median))
high_sugar = liquor_df.query('residual_sugar >= {}'.format(median))

# get the mean for quality
mean_quality_low_sugar = low_sugar['quality'].mean()
mean_quality_high_sugar = high_sugar['quality'].mean()
```

```python
# create a bar chart with proper labels
locations = [1,2]
heights = [mean_quality_low_sugar, mean_quality_high_sugar]
labels = ['Low', 'High']

plt.bar(locations, heights, tick_label = labels)
plt.title('Average quality ratings by residual sugar')
plt.xlabel('Residual sugar')
plt.ylabel('Average quality rating');
```

```python
# use groupby to get the mean quality for each acidity levels
acidity_level_mean = liquor_df.groupby('acidity_levels')['quality'].mean()
acidity_level_mean
```

```python
# create a bar chart with proper labels
locations = [2,3,4,1]
heights = acidity_level_mean
labels = ['low', 'medium', 'moderately high', 'high']

plt.bar(locations, heights,  tick_label = labels)
plt.title('Average quality ratings by acidity level')
plt.xlabel('Acidity level')
plt.ylabel('Acidity quality rating');
```
